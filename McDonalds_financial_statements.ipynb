```python
# mini project on McDonalds Financial Statements

from google.colab import files
uploaded = files.upload()
import pandas as pd

import io
name = list(uploaded.keys())[0]
data = pd.read_csv(io.BytesIO(uploaded[name]))
data.head()
```

```python
# 1. Identify Dataset and Understand It

'''Total entries: 21 rows
(corresponding to 21 years, from 2002 to 2022).

Columns: 17
features including financial metrics such as Market Cap, Revenue, Earnings, P/E Ratio, and Total Assets.

Data Types: Mostly numerical (float64), with the year column being an integer (int64).'''
```

```python
# 2. Detail explanation of dataset with feature and source of the data.

'''Year: Year of the financial data (2002â€“2022).
Market Cap: Market capitalization in billions of dollars.
Revenue: Total revenue in billions.
Earnings: Total earnings in billions.
P/E Ratio: Price-to-Earnings ratio.
Operating Margin: Operating margin percentage.
EPS (Earnings per Share): Earnings per share in dollars.
Shares Outstanding: Number of shares outstanding in billions.
Cash on Hand: Cash reserves in billions.
Dividend: Stock split adjusted dividends.
Total Assets: Total assets in billions.
Total Debt: Total debt in billions.
Total Liabilities: Total liabilities in billions.

source of the data was kaggle'''
```

```python
# 3. Prepare code snippet that has the exploration of your dataset(s) through carefully
# crafted interactions and visualizations.

import matplotlib.pyplot as plt
import seaborn as sns
new_data = data
plt.figure(figsize=(14, 8))
sns.histplot(new_data['Market cap ($B)'], kde=True, color='blue', label='Market Cap', bins=10)
sns.histplot(new_data['Revenue ($B)'], kde=True, color='green', label='Revenue', bins=10)
sns.histplot(new_data['Earnings ($B)'], kde=True, color='red', label='Earnings', bins=10)

plt.title("Distribution of Key Financial Metrics")
plt.xlabel("Value ($B)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 4. Perform IDA and EDA Approach.
# IDA
print("Dataset Shape:", data.shape)

print("Data Types:\n", data.dtypes)

print("Missing Values:\n", data.isnull().sum())

print(data.describe())

data_cleaned = data.dropna()

data_filled = data.fillna(data.mean())

#EDA
# Histogram for Market Cap
sns.histplot(data['Market cap ($B)'], kde=True)
plt.title("Distribution of Market Cap")
plt.show()

# Box plot for Earnings
sns.boxplot(data['Earnings ($B)'])
plt.title("Box Plot of Earnings")
plt.show()

# Scatter plot between Revenue and Market Cap
sns.scatterplot(x=data['Revenue ($B)'], y=data['Market cap ($B)'])
plt.title("Revenue vs Market Cap")
plt.show()
```

```python
# 5. Demonstrate a nuanced understanding of the important features of the dataset. High-
# level insights (important descriptive information, major trends, notable outliers, etc.)
# should be prominent in your resource. Statistical analyses may be included if appropriate.

market_cap_mean = data['Market cap ($B)'].mean()
market_cap_std = data['Market cap ($B)'].std()
print(f"Market Cap Mean: ${market_cap_mean}B, Standard Deviation: ${market_cap_std}B")

revenue_mean = data['Revenue ($B)'].mean()
revenue_std = data['Revenue ($B)'].std()
print(f"Revenue Mean: ${revenue_mean}B, Standard Deviation: ${revenue_std}B")

earnings_mean = data['Earnings ($B)'].mean()
earnings_std = data['Earnings ($B)'].std()
print(f"Earnings Mean: ${earnings_mean}B, Standard Deviation: ${earnings_std}B")

op_margin_mean = data['Operating Margin (%)'].mean()
op_margin_std = data['Operating Margin (%)'].std()
print(f"Operating Margin Mean: {op_margin_mean}%, Standard Deviation: {op_margin_std}%")
```

```python
# 6. Provide data visualization to the end user.

# Distribution of Key Financial Metrics
plt.figure(figsize=(12, 6))
sns.histplot(new_data['Market cap ($B)'], kde=True, color='blue', label='Market Cap', bins=10)
sns.histplot(new_data['Revenue ($B)'], kde=True, color='green', label='Revenue', bins=10)
sns.histplot(new_data['Earnings ($B)'], kde=True, color='red', label='Earnings', bins=10)

plt.title("Distribution of Market Cap, Revenue, and Earnings")
plt.xlabel("Value ($B)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()

# Time Series Analysis: Market Cap, Revenue, and Earnings Over Time
plt.figure(figsize=(14, 8))
sns.lineplot(x=new_data['Year'], y=new_data['Market cap ($B)'], label='Market Cap', color='blue')
sns.lineplot(x=new_data['Year'], y=new_data['Revenue ($B)'], label='Revenue', color='green')
sns.lineplot(x=new_data['Year'], y=new_data['Earnings ($B)'], label='Earnings', color='red')

plt.title("Market Cap, Revenue, and Earnings Over Time (2002â€“2022)")
plt.xlabel("Year")
plt.ylabel("Value ($B)")
plt.legend()
plt.grid(True)
plt.show()
```
